```python
import os
import json
import pandas as pd
from pandas.io.json import json_normalize
from pprint import pprint

with open('Priorities_export.json') as data_file:
	data = json.load(data_file)


def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```
